```python
from abc import ABC, abstractmethod
from typing import TypeVar, Generic

T = TypeVar('T')

class BaseCoding(Generic[T], ABC):
    @abstractmethod
    def encode(self, x:T)->T:
        pass
    @abstractmethod
    def decode(self, x:T)->T:
        pass

class TextCoding(BaseCoding[str]):
    def encode(self, x:str)->str:
        return x
    def decode(self, x:str)->str:
        return x
```

```python
from typing import List

class ImageCoding:
    def encode(self, x:List[str])->List[str]:
        return x;
    def decode(self, x:List[str])->List[str]:
        return x;

BaseCoding.register(ImageCoding)

imageCoding = ImageCoding()
print(imageCoding.encode(["hello"]))
```

```python
offer = 'yes'
if offer == 'yes':
    print(imageCoding.decode(["new career in enterprise"]))
elif offer == 'no':
    print(imageCoding.decode(["new career in outside"]))
else:
    print(imageCoding.decode(["new career is myself"]))
```

```python
class Monitor:
    def __init__(self):
        self._resource = []
    @property
    def resource(self):
        return self._resource

    @resource.setter
    def resource(self, resource):
        self._resource = resource

monitor = Monitor()
monitor.resource = list(range(100))
print(monitor.resource)
```
