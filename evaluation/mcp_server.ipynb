```python
from fastmcp import Client
import asyncio
import os

from dotenv import load_dotenv

load_dotenv()

BAIDU_API_KEY = os.getenv("BAIDU_API_KEY")
BAIDU_SEARCH = f"http://appbuilder.baidu.com/v2/ai_search/mcp/sse?{BAIDU_API_KEY}"

async def main():
    async with Client(BAIDU_SEARCH) as client:
        tools = await client.tools()
        print("Tools", tools)

asyncio.run(main())
```
