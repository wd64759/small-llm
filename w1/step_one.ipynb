```python
class LinkNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def insert(head:LinkNode, pos:int, val):
    if pos == 0 or not head:
        new_node = LinkNode(val)
        new_node.next = head
        return new_node
    else:
        head.next = insert(head.next, pos-1, val)
        return head
    
head = LinkNode(1)
n = insert(head, 5, 2)
n
```

```python
def AND(s1, s2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = w1*s1 + w2 * s2
    if tmp <= theta:
        return 0
    else:
        return 1
    
print("AND(0,0)", AND(0,0))
print("AND(1,0)", AND(1,0))
print("AND(0,1)", AND(0,1))
print("AND(1,1)", AND(1,1))

def NAND(s1, s2):
    w1, w2, theta = -0.5, -0.5, -0.7
    tmp = w1*s1 + w2 * s2
    if tmp <= theta:
        return 0
    else:
        return 1
    
print("NAND(0,0)", NAND(0,0))
print("NAND(1,0)", NAND(1,0))
print("NAND(0,1)", NAND(0,1))
print("NAND(1,1)", NAND(1,1))

```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y


print("XOR(0,0)", XOR(0,0))
print("XOR(1,0)", XOR(1,0))
print("XOR(0,1)", XOR(0,1))
print("XOR(1,1)", XOR(1,1))


```
